# Module: Communicating Analysis Results
## Sprint: Visualising Data using Tableau
## Part 2: Making dashboards more interactive


Let’s go back to our theory book Storytelling with Data. Through chapters 4 to 6 you will learn more about focusing your audience attention, thinking like a designer and dissecting your visuals. After theory we will explore options on how you can make dashboards more interactive. The video No 4 from “Alex The Analyst” channel will teach you how to use joins in Tableau and 5th video will guide you on your first project on Tableau. But do not worry, you will get a bit more hands-on experience in our Hands On and Graded tasks. Next we will deep dive into more specific topics. We will start with filters, covering basic Tableau filters, action filters and parameter filters which allow filtering across multiple data sources.
